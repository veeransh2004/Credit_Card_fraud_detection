```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('creditcard.csv')


valid_transactions = data[data['Class'] == 0]
fraudulent_transactions = data[data['Class'] == 1]

# Number of fraudulent and valid transactions
num_fraudulent = len(fraudulent_transactions)
num_valid = len(valid_transactions)

print("Number of fraudulent transactions:", num_fraudulent)
print("Number of valid transactions:", num_valid)

# Maximum, minimum, and mean amounts for valid transactions
max_valid_amount = valid_transactions['Amount'].max()
min_valid_amount = valid_transactions['Amount'].min()
mean_valid_amount = valid_transactions['Amount'].mean()

print("\nValid Transactions:")
print("Maximum amount for valid transactions:", max_valid_amount)
print("Minimum amount for valid transactions:", min_valid_amount)
print("Mean amount for valid transactions:", mean_valid_amount)

# Maximum, minimum, and mean amounts for fraudulent transactions
max_fraudulent_amount = fraudulent_transactions['Amount'].max()
min_fraudulent_amount = fraudulent_transactions['Amount'].min()
mean_fraudulent_amount = fraudulent_transactions['Amount'].mean()

print("\nFraudulent Transactions:")
print("Maximum amount for fraudulent transactions:", max_fraudulent_amount)
print("Minimum amount for fraudulent transactions:", min_fraudulent_amount)
print("Mean amount for fraudulent transactions:", mean_fraudulent_amount)

# Plotting the correlation matrix with enhanced settings
correlation_matrix = data.corr()
# Plotting the correlation matrix with modified settings
plt.figure(figsize=(12, 10))  # Enlarge the plot size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 8})
plt.title('Correlation Matrix of Credit Card Data', fontsize=14)
plt.tight_layout()  # Adjust layout to prevent cutting off labels
plt.show()
```
